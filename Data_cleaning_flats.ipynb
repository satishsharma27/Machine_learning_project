```python
def treat_price(x):
    # Check if x is a float, return as is if true
    if isinstance(x, float):
        return x
    
    # Check if x is a list and contains at least two elements
    if isinstance(x, list) and len(x) >= 2:
        # Check if the second element is 'Lac'
        if x[1] == 'Lac':
            return round(float(x[0])/100, 2)  # Convert Lac to Crore
        else:
            return round(float(x[0]), 2)  # For Crore values
    
    # If x is not as expected, return NaN or handle as needed
    return float('nan')
```

```python
import re
def clean_price(price):
    if pd.isna(price):
        return None 
    
    match = re.search(r'₹\s?([\d,]+)/sq\.ft\.', price)
    if match:
        return float(match.group(1).replace(',',''))
    return None
```
